### Stack Plots
- A stack plot is a graph where multiple data series are stacked on top of each other. 

- Each layer shows the individual category and the top of line shows the cumulative total (how each individual part contributed to total over time).
- Use stack plot when you want to:
    1. Show Cumulative data over time (Cumulative data over time means data that keeps adding together as time passes)
    
    2. Compare multiple components
    3. Highlight trends of individual parts and the whole

```python
import matplotlib.pyplot as plt

days = ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
direct_marketing = [50, 60, 70, 80, 90, 100, 110,]
organic_marketing = [30, 40, 50, 55, 60, 70, 80]
social_marketing = [20, 25, 30, 35, 40, 50, 60]

# To present one day data we can use normal graphs or plots, for example here pie chart is suitable for one day data.
# plt.pie([direct_marketing[0],organic_marketing[0],social_marketing[0]], labels=["Direct Marketing","Organic Marketing","Social Marketing"],autopct="%1.1f%%")

# To present each days data we need to use stack plot
plt.stackplot(days,direct_marketing,organic_marketing,social_marketing,labels=["Direct Marketing","Organic Marketing","Social Marketing"])
plt.title("Marketing data for the week")
plt.xlabel("Days")
plt.ylabel("No.of Customers")
plt.legend()
plt.grid()
plt.show()
```
