### Histograms
- A histogram groups numbers into ranges (called bins) and shows how many values fall in each range using bars.

- Use histogram when:
    1. You want to show the distribution of continuous data.
     
    2. Working with large datasets to summarize trends. 
    3. You need to look at the graph to understand how often values occur, how the data is spread, whether it leans to one side, and if there are any patterns.

```python
import matplotlib.pyplot as plt

stu_scored = [40, 45, 50, 52, 55, 60, 62, 65, 70, 75]  
# plt.hist(stu_scored, bins=4, color='purple', edgecolor='black')  
# Bins divide the data into groups, and the histogram counts how many values fall in each group.
# Bins = 4 means divide the data into 4 groups.

# How do we know the range of bins?
# Bin = Max Value - Min Value / Number of Bins
# Example: 75-40/4 => 8.75
# So the range of each groups will be 8.75 i.e for first bin 40 - 48.75 and second bin 48.75 - 57.5 and so on.

# OR

# If bins is passed as a list, then you are manually defining the bin ranges instead of letting Matplotlib calculate them automatically.
bin = [20,30,40,50,60,70,80]
plt.hist(stu_scored, bins=bin, color='purple', edgecolor='black')  
# Now ranges would be defined like this 20-30, 30-40,40-50 and so on.

plt.xlabel("Marks")  
plt.ylabel("Number of Students")  
plt.title("Histogram of Student Marks")  

# plt.grid(True) #Show Grid Lines
# plt.grid(False) # Hide Grid Lines

# plt.grid(True,axis="y") # horizontal grid lines only
# plt.grid(True,axis="x") # vertical grid lines only
# plt.grid(True,axis="both")

plt.grid(True,axis="y",linestyle="--",alpha=.7)
plt.show() 
```

In a histogram, bins usually follow this rule:<br>

Left side = inclusive (included)<br>
Right side = exclusive (not included)<br>
Last bin includes both ends

If bins are:<br>
bins = [40, 50, 60, 70, 80]

The ranges become:<br>
Bin 1 - 40 ≤ x < 50 <br>
B2 - 50 ≤ x < 60<br>
B3 - 60 ≤ x < 70<br>
B4 - 70 ≤ x ≤ 80

```python
# Example 2 : Multiple Datasets
legit_transactions = [
    2.99, 5.49, 8.99, 12.50, 14.99, 19.99, 23.45, 29.99, 34.99, 39.50,
    45.00, 49.99, 55.25, 60.00, 75.99, 89.99, 120.50, 150.00, 199.99, 
    249.99, 300.75, 450.00, 600.00, 850.00, 1200.00
]

fraud_transactions = [
    50,100,150,200,300,500,500,750,1000,1000,
    1200,1500,1500,2000,2500,3000,300
]

plt.hist(fraud_transactions,color='red',bins=10,label="Fraud Transaction",edgecolor="black",alpha=.5 )
plt.hist(legit_transactions,color='green',bins=10,label="Legit Transaction",edgecolor="black",alpha =.5 )
plt.title("Transaction Amount Distribution - Legit Vs Fraud")
plt.legend()
plt.xlabel("Transactions Amount (in INR)")
plt.ylabel("Number of Transactions Done")
```

**axvline**<br>
axvline is used to draw a vertical reference line on a plot, common to mark cutoffs, averages or pass marks.

```python
# Example 3 axvline
stu_marks = [40, 45, 50, 52, 55, 60, 62, 65, 70, 75, 20, 15, 5, 40, 25, 30, 32, 33]  
plt.hist(stu_marks, bins=7, color='purple', edgecolor='black')
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.axvline(
    x=33,
    label="Passing Marks (33)",
    color="green",
    linewidth="2",
    linestyle="--"
)
plt.legend()
plt.show()  
```
