### Weekply Temperature
- Compare temperature of cities over a week using subplots & loops.

```python
days = ["Mon","Tue","Wed","Thu","Fri"]
cities = ["New York","London","Delhi","Tokyo"]

temperatures = [
    [22,23,21,24,25], # New York
    [18,19,17,20,21], # london
    [30,32,31,33,34], # Delhi
    [25,26,24,27,28] # Tokyo
]

import matplotlib.pyplot as plt

figure, axes = plt.subplots(2,2)
k = 0

for i in range(2):
    for j in range(2):
        axes[i][j].plot(days,temperatures[k],marker="o")
        axes[i][j].set_title(cities[k])
        axes[i][j].grid()
        k += 1

figure.suptitle("Weekly Temperature of 4 Cities")
figure.supxlabel("Days")
figure.supylabel("Temperatures")
figure.tight_layout()
```
