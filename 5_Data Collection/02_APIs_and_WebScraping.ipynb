## Working with `requests`
- The `requests` module in Python is a library that lets us send HTTP requests to websites or APIs. 

- We can - get, post, download, and send data over the internet easily.
- To install it - `pip install requests`
- use `toscrape.com` or `scrapethissite.com` website for performing web scraping activities.

### Basic GET Request - used to retrieve a webpage or API response.

```python
import requests

# Using scrapethissite.com > countries 
url = "https://www.scrapethissite.com/pages/simple/"
res = requests.get(url)

# print(res.status_code)
# print(res.text) #  HTML content of the page as a string
# print(res.headers) # Metadata like content type, server info etc.

# Status Code ------
# 200 - OK
# 301/302 - Redirect
# 400 - Bad Request
# 401 - Unauthorized
# 403 - Forbidden
# 404 - Not Found
# 500 - Server Error
```

#### As a Data Scientist, it is a good practice to ALWAYS check for status_code before working with data.

#### When doing web scraping, we create a separate folder to download and store different web files. This helps keep the scraped data organized and easy to manage.

```python
if res.status_code == 200:
    with open("scraped_data/data1.html","w") as f:
        f.write(res.text)
else:
    print("Failed to fetch webpage data.")
```

### Note - 
1. Just like GET requests, we also have POST, DELETE & other requests which are out of scope for our course but GET is going to be the most relevant for us for now (in context of web scraping).

2.  We can also use `try-except` blocks to handle exceptions in our code.

```python
try:
    r = requests.get("https://www.scrapethissite.com/pages/simple/",timeout=5)
    r.raise_for_status()  # checks the HTTP response status code and raises an error if the request failed with standard name/message.
    print("Request successful")
except requests.exceptions.RequestException as e:
    print("Error:",e)
```

## Working with BeautifulSoup
- BeautifulSoup (bs4) is a Python library used to parse HTML and XML documents. 
- It helps you extract data from webpages after downloading them with `requests`.
- It creates a parse tree from the data from which we can extract text, attributes, links, tables etc.

#### Install & Import

`from bs4 import BeautifulSoup`

#### For faster parsing we will install

```python
from bs4 import BeautifulSoup

with open("scraped_data/data1.html","r") as f:
    html_content = f.read()

soup = BeautifulSoup(html_content,"lxml")
```

```python
# soup
```

From the above file - data1.html we want to gather a information of countries like it's name and population in tabular form. 

**Methods in BeautifulSoup**
- `find()` - find 1st matching element

- `find_all()` 
- `find_parent()` 
- `find_next()` 
- `find_previous()` 

Note: To anaylze the elements and it's css attributes use browser inspect tool.

```python
# find 1st matching element - both does same thing
soup.find("h3")
soup.h3
```

```python
all_h3 = soup.find_all("h3")
for h3 in all_h3:
    # print(h3.get_text(strip=True))
    name = h3.get_text(strip=True)
    print(name)
    
    # get_text() - Gets all the text inside the <h3> tag
    # strip=True - Removes extra spaces, tabs, and new lines from the beginning and end
```

```python
for h3 in all_h3:
    print(h3.find_parent("div")["class"])
```

**CSS Selectors**
- `select()`

- `select_one()`

```python
for h3 in all_h3:
    # print(h3.find_next("div")) 
    # print(h3.find_next("div").select("span.country-population"))
    # print(h3.find_next("div").select("span.country-population")[0])
    popluation = (h3.find_next("div").select("span.country-population")[0].get_text(strip=True)) #country-population is a css attribute exists in the span element
    print(popluation)
```


**Attributes**
- `get()` - gets a specific attribute value

- `.attrs` - gives all attributes including (class `.` & id `#`) of the tag in dictionary form

```python
for h3 in all_h3:
    name = h3.get_text(strip=True)
    population = h3.find_next("div").select("span.country-population")[0].get_text(strip=True)

    print(h3.get("class"))
    # print(h3.get("id"))
```

```python
for h3 in all_h3:
    name = h3.get_text(strip=True)
    population = h3.find_next("div").select("span.country-population")[0].get_text(strip=True)

    print(h3.attrs)
```

Now we are extracting details of countries and saving it in a list and converting it into a dataframe.

```python
all_Countries = []
for h3 in all_h3:
    name = h3.get_text(strip=True)
    population = h3.find_next("div").select("span.country-population")[0].get_text(strip=True)
    all_Countries.append([name,population])
all_Countries

import pandas as pd
df = pd.DataFrame(all_Countries,columns=["Name","Population"])
```

Once we get a cleaned data we stored it inside the file.

```python
df.to_csv("cleaned_data/data.csv",index=False)
```
