### Seaborn
- Just like Matplotlib, Seaborn is a Python library used for data visualization.

- It is developed using Matplotlib as its foundation.
- Use Matplotlib when:
    - You need full control over the plot.

    - You need highly customized graphs.
    - You are building plots from scratch.
- Use Seaborn when:
    - You want more attractive and modern statistical plots.

    - You are working with data analysis or data science.
    - You want better default styles and colors.
- Sometimes both Seaborn and Matplotlib are used together.
- Benefits of using seaborn:
    1. It provides inbuilt datasets, which are helpful for learning and allow quicker visualization. The availability of built-in datasets helps learners to learn more quickly.
    
    2. It provides high-level features such as automatic styling and many more.
    3. Seaborn works very well with Pandas DataFrames. Unlike Matplotlib, where working directly with DataFrames is not always easy, Seaborn can use Pandas DataFrames directly, which makes creating visualizations simpler and faster.

- To install seaborn: `pip install seaborn`
