##### Mathematical Functions
- NumPy is massive & provides a wide range of built-in mathematical functions that are highly optimized and can operate element-wise on arrays.

```python
# Aggregate Functions
# Functions that perform a calculation on a set of values and return a single summary value.

import numpy as np
arr = np.array([1,2,3,4,5])

print(np.sum(arr))
print(np.prod(arr))
print(np.min(arr))
print(np.max(arr))
print(np.argmin(arr)) # returns index of min value 
print(np.argmax(arr)) # returns index of max value 
print(np.mean(arr))
print(np.median(arr))
print(np.std(arr))
print(np.var(arr))
```

```python
# Power functions
# Power functions are functions that raise numbers (or arrays) to a certain exponent.

arr2 = np.array([1,2,3,4,5])
print(np.square(arr2)) 
print(np.sqrt(arr2)) # returns square root of all elements
print(np.pow(arr,3)) #  returns a raise to the power b (a here is arr elements)
```

```python
# Log & Exponential Functions
print(np.log(arr2))  
print(np.log10(arr2))  
print(np.log2(arr2))  
print(np.exp(arr2)) 
```

```python
# Rounding Functions
print(np.round(2.578)) # rounds off to nearest value 
print(np.round(2.478)) # rounds off to nearest value
print(np.ceil(4.278)) # rounds up
print(np.floor(4.999)) # rounds down
print(np.trunc(2.678)) # truncates (removes) the fractional or decimal part
```

```python
# Miscellaneous Functions
arr = np.array([1, 2, -5, 3, 8, -4, 2, 5]) 

print(np.abs(arr)) # returns absolute value
print(np.sort(arr)) # returns sorted(arranges in increasing order) values
print(np.unique(arr)) 

# There are many other built-in functions that we will use and learn about in later chapters. 
```
