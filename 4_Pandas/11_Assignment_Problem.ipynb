B. Select only those rows where:
- petal_length > 4.5
- species = "Iris-virginica"

```python
print(df[df["petal_length"]>4.5])
print(df[df["species"]=="Iris-virginica"])

# Combining both conditions together
print(df[(df["petal_length"]>4.5) & (df["species"]=="Iris-virginica")])
```

C. Group by species and compute:
- Average sepal_length
- Maximum petal_width
- Standard deviation of sepal_width

```python
df.groupby("species")["sepal_length"].mean()
```

```python
df.groupby("species")["petal_width"].max()
```

```python
df.groupby("species")["sepal_width"].std()
```

```python
# Combining 3 conditions together
df.groupby("species").aggregate({
    "sepal_length":"mean",
    "petal_width":"max",
    "sepal_width":"std"
})
```

D. Create a new column: `"petal_ratio" = petal_length/petal_width`
- For each species, find the average petal_ratio.


```python
df = df.assign(petal_ratio = df["petal_length"]/df["petal_width"])
```

```python
df
```

```python
df.groupby("species")["petal_ratio"].mean()
```
