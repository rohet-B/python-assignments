### Question 1:
For the following dataset generated using NumPy:<br><br>
   `np.random.normal(loc=70,scale=10,size=200)`
- Plot a histogram of scores
- Draw a vertical line for the mean & try to interpret the results.

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.normal(loc=70,scale=10,size=200)
# print(data)

mean_value = np.mean(data)

# Histogram
plt.hist(data, bins=10, color='purple', edgecolor='black')  

# Vertical line
plt.axvline(mean_value,color="red",linestyle="dashed",linewidth=2)

plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.title("Histogram of Scores with Mean Line")
plt.show()
```

### Question 4:
For the seaborn `tips` dataset, do the following:
- Create a scatter plot of `total_bill` vs `tip`
- Color points by `sex`
- Use different markers for `smoker`
- Add a title and axis labels
- Analyze: Do Smokers tend to tip differently ?

```python
sns.scatterplot(
    data = tips,
    x = "total_bill",
    y = "tip",
    hue = "sex",
    style="smoker"
)
plt.title("Total_bill vs tip")
plt.xlabel("Total bill")
plt.ylabel("Tips")
```

### Question 5: 
For the seaborn `tips` dataset, do the following:
- Create a pivot table of average tip
    - Rows: `day`
    - Columns: `time`
- Plot the pivot table as a `heatmap`
- Annotate values
- Interpret which day - time combination performs best

```python
tips_pivot_data = tips.pivot_table(index="day",columns="time",values="tip",observed=False)
sns.heatmap(
    data = tips_pivot_data,
    cmap="coolwarm",
    annot=True, # shows numbers inside each cell
)
plt.title("Average tip per day as a Heatmap")
plt.tight_layout()
```
