### Supervised Learning
- Supervised ML trains algorithms on labeled datasets, meaning each input has a known correct output, to learn patterns and predict outcomes for new data.

### Meaning of x, y and features.
- x (small x) represents a single input 

- X (capital X) represents multiple inputs 
- y (small y) represents a single output
- Y (Capital Y) represents multiple outputs

- In Machine Learning, features are the input variables (columns) used to make predictions.
    - Input columns are called independent features.

    - Output columns are called dependent features.


### Process of creating Machine Learning Model
1. Data Preprocessing - Data Cleaning + Feature Engineering

2. Selecting a Machine Learning Model Algorithm
3. Train & Test the Model
4. To evaluate if the model is experiencing underfitting or overfitting.

### Types of problems we going to solve in Supervised Learning

1. `Regression` - Output is a continuous numerical value.
    - There are 2 types of regression:
        1. `Linear Regression` - Uses one input variable (x) to predict output (y). Example: Predict salary based on experience

        2. `Multiple Linear Regression` - Uses more than one input variable (X₁, X₂, …) to predict single output (y). Example: Predict salary based on experience, education and skills


2. `Classification` - Output is a category or class.
    - There are 3 types of classification:
        1. `Binary Classification` - Only 2 possible outputs and only one output is selected.

        2. `Multi-class Classification` - More than 2 possible outputs, but only one output is selected.
        3. `Multi-label classification` - Multiple labels can be assigned to a single input. For example: One movie can be Action + Comedy + Drama. It means more than one output is possible and multiple outputs can be selected at the same time.

### Linear Regression
1. It is a supervised ML algorithm for regression problems.

2. Linear Regression (LR) explains the relationship between a dependent variable (output) and one or more independent variables (inputs) by drawing the best straight line through the data.
3. Linear Regression is an algorithm that predicts the relationship between x and y by finding the *best-fit line*. 


### What is best-fit line?
- The best-fit line is the straight line that passes as close as possible to all the data points.

- Best-fit line also known as Regressor or Hypothesis.
- Best-fit line represents the predicted values.
- The gap from `actual value` to `predicted value` or difference between `actual value` and `predicted value` is known as `error in prediction` or `residual error`.
- Many lines can be drawn on a graph, but the best-fit line is the one with the least error in prediction or least gap between the actual and predicted value.


- The equation for this line is `y=mx+c` but in Machine Learning it is written with different notations: `hθ(x) = θ₀ + θ₁x₁`
    - Where: 
        - `hθ(x)` means hypothesis function or our model
        - `θ₀ ` means intercept or Bias
        - `θ₁` means slope or coefficient of x or weights
        - `x₁` means 1st feature
    - When we have multiple independent features, it is called multiple linear regression:-
        - `hθ(x) = θ₀ + θ₁x₁ + θ₂x₂ + … + θₙxₙ`

### Cost Function
- Now the question is how to find the best-fit line.
The best-fit line is the one with the smallest gap between actual and predicted values. To measure this gap, we use a cost function.

- A cost function is a formula that tells us how wrong our model’s predictions are.

- In simple terms: It measures the error between the actual value (y) and the predicted value (ŷ).
- Most common Cost Function used for Linear Regression is `Mean Squared Error` (MSE) :-<br>
