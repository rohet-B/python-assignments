#### Let's See which one is fast Python List or NumPy array 

```python
# Execution performance
import time
size = 1_000_000

# Calculating square of list items
# python list
py_list = list(range(size)) # [1,2,3,4,........]
start = time.time() #prints current time
sq_list = [x**2 for x in py_list]
end = time.time() 
print(f"Python list time = {end-start} seconds")

# NumPy arrays
np_arr = np.array(py_list)
start = time.time()
sq_array = np_arr ** 2
end = time.time()
print(f"Python list time = {end-start} seconds")
```

```python
# Memory Usage Comparison of Python list & NumPy array.
import sys

total_size = sys.getsizeof(py_list)
# sys.getsizeof(object) is a Python built-in function that tells How many bytes of memory this object is using right now.

for item in py_list:
    total_size += sys.getsizeof(item) 
    # Since we know that Python lists are stored in contiguous memory, the real elements are not stored directly. Instead, the list stores pointers (references) that point to the actual elements, which are located elsewhere in memory.
# This is why sys.getsizeof(py_list) only measures the size of the list structure (the pointers), not the memory used by the actual elements.
# Each element (like an integer, string, etc.) is a separate object in memory, so you would need to add their sizes individually to get the total memory used.

print(f"Python list size = {total_size} bytes.") 
print(f"NumPy array size = {np_arr.nbytes} bytes. ")
```
