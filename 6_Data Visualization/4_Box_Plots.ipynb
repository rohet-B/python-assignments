### Box Plots
A box plot (also called a box-and-whisker plot) is a graph used to show how data is distributed or spread using five key values.
“How the data is distributed” means how the values are spread or arranged across different ranges.
1. Minimum - Smallest Non Outlier Value 

2. Q1 (First Quartile) - 25% of data
3. Q2 (Median) - Middle Value
4. Q3 (Third Quartile) - 75% of data
5. Maximum - Largest Non Outlier Value

It can also show **outliers**.

### Inner working of boxplot
Given Data - 7, 8, 5, 6, 9, 4, 10, 12, 15<br>
1. Sort the provided data - `4,5,6,7,8,9,10,12,15`

2. Finding Q2 Median (Middle Value):
    -  If no.of elements in data are odd:<br>
        => Take the middle value for example here is 8 <br>
        `4,5,6,7,  8,  9,10,12,15`

    -  IF no.of elements in data are even:<br>
        Find Average of 2 middle numbers: 5,10,15,20<br>
        `10 + 15 / 2` => 12.5
3. Now Finding Q1 and Q2 
    - For Odd no.of Elements:
        - Now Consider first half of the data including median<br>

            `4, 5, 6, 7, 8`
        - Now Find median in it:<br>
            `4, 5,    6,     7, 8`
        - The median value here is 6 and it is known as Q1 Value (25% of Data)
        - Now Do the same with Second Half of data including median
    - For Even no.of Elements:
        - Suppose there are 6 elements: `4, 5, 6, 7, 8, 9`

        - There median or Q2 is 6.5
        - First Half - 4, 5, 6
        - Second Half - 7, 8, 9
        - Find Medians in both half 
            - Q1 = 5
            - Q3 = 8
4. Now Finding Minimum & Maximum Value
    - Minimum Value Also Called Lower Whisker
        - Formula is: 
            `lower = Q1 - (1.5 * IQR)`

    - Maximum Value Also Called Upper Whisker
        - Formula is: 
            `upper = Q3 + (1.5 * IQR)` 
    - Where IQR Stands for Inter- Quartile Range and it's formula is `Q3-Q1` and `1.5` is whisker length.
5. Note:
    - If the calculated lower and upper whisker limits are 0 and 16, the box plot will not show exactly 0 and 16.
    - The Graph only represents data values that lie between 0 and 16. 
    - Any values less than 0 or greater than 16 are called outliers and are shown separately in the plot.

```python
import matplotlib.pyplot as plt
data = [7, 8, 5, 6, 9, 4, 10, 12, 15]

plt.boxplot(data) # Vertical Box Plot
# plt.boxplot(data,vert=False) # Horizontal Box Plot

plt.grid()
plt.show()
```

### Additional Operations on Box Plots
1. Multiple Datasets on a single Box Plot

```python
data1 = [7, 8, 5, 6, 9, 7, 8, 10, 4, 6]
data2= [5, 6, 7, 8, 5, 4, 6, 7, 5, 6] 
plt.boxplot([data1,data2],tick_labels=["Dataset 1", "Dataset 2"])
plt.grid()
plt.show() # Each box represents a different dataset.
```

2. Showing Mean
- `showmeans=True` - adds the mean marker

- `meanline=True` - draws a line instead of a point

```python
data1 = [7, 8, 5, 6, 9, 7, 8, 10, 4, 6]
plt.boxplot(data1, showmeans=True, meanline=True)
plt.grid()
plt.show() 
```

### Use boxplot when you want:
1. To understand how the data is distributed. “How the data is distributed” means how the values are spread or arranged across different ranges.

2. To identify unusual values (outliers) in the dataset.
3. To compare how data is distributed across different groups.
