### Supervised Learning
- Supervised ML trains algorithms on labeled datasets, meaning each input has a known correct output, to learn patterns and predict outcomes for new data.

### Meaning of x, y and features.
- x (small x) represents a single input 

- X (capital X) represents multiple inputs 
- y (small y) represents a single output
- Y (Capital Y) represents multiple outputs

- In Machine Learning, features are the input variables (columns) used to make predictions.
    - Input columns are called independent features.

    - Output columns are called dependent features.


### Process of creating Machine Learning Model
1. Data Preprocessing - Data Cleaning + Feature Engineering

2. Selecting a Machine Learning Model Algorithm
3. Train & Test the Model
4. To evaluate if the model is experiencing underfitting or overfitting.

### Types of problems we going to solve in Supervised Learning

1. `Regression` - Output is a continuous numerical value.
    - There are 2 types of regression:
        1. `Linear Regression` - Uses one input variable (x) to predict output (y). Example: Predict salary based on experience

        2. `Multiple Linear Regression` - Uses more than one input variable (X₁, X₂, …) to predict single output (y). Example: Predict salary based on experience, education and skills


2. `Classification` - Output is a category or class.
    - There are 3 types of classification:
        1. `Binary Classification` - Only 2 possible outputs and only one output is selected.

        2. `Multi-class Classification` - More than 2 possible outputs, but only one output is selected.
        3. `Multi-label classification` - Multiple labels can be assigned to a single input. For example: One movie can be Action + Comedy + Drama. It means more than one output is possible and multiple outputs can be selected at the same time.
