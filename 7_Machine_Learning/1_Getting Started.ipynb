#### Machine Learning
- ML is a part of computer science where we teach computers to learn from data, find patterns and make decision or predictions.

### ML Types
1. `Supervised Learning`
    - Contains Labelled data, labelled data is raw data such as images, text or audio that has been paired with a meaningful tag, label or annotation that informs a machine learning model about the desired outcome.

    - It mainly solves two types of problems:
        1. `Classification` - It means predicting a category or class. The output is discrete (fixed labels). Example: If you train a model with emails labeled as spam/not spam, it learns patterns and then classifies new emails with these categories.

        2. `Regression` - It means predicting a continuous value (number). The output is numeric (real values). If you train a model with data of house size and prices, it can predict the price of a new house.
    

2. `Unsupervised Learning` 
    - Contains unlablled data.

    - The model tries to find patterns or structure on its own.

    - It mainly solves one problem:
        1. `Clustering` - It means grouping similar data points together. There are no predefined labels. The model automatically creates groups (clusters). Example: Suppose you have data of students: Some scores high, Some average, Some low. Without giving labels, the model will automatically group them into: High Performers, Average Performers and Low Performers.

3. `Reinforcement Learning`
    - Reinforcement Learning (RL) is a type of machine learning where a model (called an agent) learns by interacting with an environment and receiving rewards or penalties.

    - Reinforcement Learning = Learning by doing + feedback (reward/penalty)

    - Example: 
        A self-driving car learns how to drive using reinforcement learning. <br><br>Agent → the car<br> Environment → road<br> traffic signals<br> Action → accelerate, brake, turn <br><br> Reward → <br>+10 for staying in lane <br>+20 for reaching destination <br> -10 for going off-road <br> -50 for accident <br><br> The car tries different actions again and again.<br> It learns which actions give more reward and avoids bad ones.<br> Over time, it becomes a safe driver.
