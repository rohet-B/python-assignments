### Question 1 - Start by downloading the zip file of [IRIS dataset](https://www.kaggle.com/datasets/arshid/iris-flower-dataset).
Display the following information
- First 10 Rows
- Shape & Data types
- Summary Statistics (mean,std,min,max)

```python
import pandas as pd
df = pd.read_csv("IRIS.csv")
```

```python
# First 10 Rows
df.head(10)
```

```python
# Shape & Data types
print(f"Shape of dataframe {df.shape}")
print(df.dtypes)
```

```python
# Summary Statistics (mean,std,min,max)
df.describe()
```

B. Select only those rows where:
- petal_length > 4.5
- species = "Iris-virginica"

```python
print(df[df["petal_length"]>4.5])
print(df[df["species"]=="Iris-virginica"])

# Combining both conditions together
print(df[(df["petal_length"]>4.5) & (df["species"]=="Iris-virginica")])
```

C. Group by species and compute:
- Average sepal_length
- Maximum petal_width
- Standard deviation of sepal_width

```python
df.groupby("species")["sepal_length"].mean()
```

```python
df.groupby("species")["petal_width"].max()
```

```python
df.groupby("species")["sepal_width"].std()
```

```python
# Combining 3 conditions together
df.groupby("species").aggregate({
    "sepal_length":"mean",
    "petal_width":"max",
    "sepal_width":"std"
})
```

D. Create a new column: `"petal_ratio" = petal_length/petal_width`
- For each species, find the average petal_ratio.


```python
df = df.assign(petal_ratio = df["petal_length"]/df["petal_width"])
```

```python
df
```

```python
df.groupby("species")["petal_ratio"].mean()
```

### Question 2 - Start by downloading the zip file of [TITANIC dataset.](https://www.kaggle.com/datasets/yasserh/titanic-dataset)

```python
df2 = pd.read_csv("Titanic-Dataset.csv")
df2
```

A) Display only columns: Name, Sex, Age, Fare, Survived.

```python
df2[["Name","Sex","Age","Fare","Survived"]]
```

B) Select passengers who:
- Are female
- Fare > 30

```python
df2[df2["Sex"]=="female"]
```

```python
df2[df2["Fare"]>30]
```

C) Group by Pclass and compute:
- Survival rate
- Average Fare
- Average Age

```python
df2.groupby("Pclass").aggregate(
   Survival_Rate = ("Survived","mean"),
   Average_Fare = ("Fare","mean"),
   Average_Age = ("Age","mean")
)
```
