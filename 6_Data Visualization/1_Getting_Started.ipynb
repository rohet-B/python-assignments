## Data Visualization
- We, as humans, process visual information faster than text or numbers, so we use Data visualization i.e. the graphical representation of data.
- The common types of plots used in data visualization are line plots, bar charts, histograms, scatter plots, box plots etc.

## Matplotlib
- Python Library used for:
    - Creating static, animated and interactive plots 
    - Low-level control over plot appearance 
    - Serving as the foundation for other libraries (Seaborn, Pandas plotting) 

- Installation:
    `pip install matplotlib`

- Visit the [Matplotlib](https://matplotlib.org/) site for more tutorials.

## Line Plots
- Type of chart used to display data points connected by straight lines, typically to show trends or changes over a continuous variable, such as time or distance.

- Use line plot when:
    - Data points are ordered 
    - You want to show progression or trends
    -  You’re comparing changes rather than individual categories 

```python
import matplotlib.pyplot as plt
```

In `matplotlib.pyplot`, Pyplot is a module of Matplotlib used to create and display graphs using simple functions.

## Plotting a single line

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(x,y,
         label="Line 1",
         color="red",
         linestyle="--",
         linewidth=2,
         marker="x",
         markersize=6)
# Line Style - Solid(-), dotted(:), Dashed(--), Dash-Dot(-.)
# Marker - Circle ("o"), Square ("s"), Triangle up ("^"), Triangle Down ("v"), Star ("*"), Plus ("+"), Cross ("x"), Dot (".")

plt.legend() # displays labels for different lines in a graph.
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Simple Line Plot")
plt.grid(True)
plt.show()
```

## Plotting multiple lines 

```python
x2 = [1,2,3,4,5]
y2 = [2,4,6,8,10]
y3 = [1,3,5,7,9]

plt.plot(x2,y,label="Line 1",color="#CBC3E3")
plt.plot(x2,y3,label="Line 2")

plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend()
# plt.show() 
# plt.savefig("line_plot.png") # format supported - png, jpg, svg, pdf
plt.show() # Used to display the graph but it also clears the graph from memory, that's why saving image after .show() can give a blank image.
```

## Format Strings in Matplotlib
- Compact way to specify the line style, marker and color in a single string.
- `'marker line color'` (Write without spaces)

```python
x3 = [1,2,3,4,5,6]
y4 = [3,6,9,12,15,18]

plt.plot(x3,y4,
         "o--r")
plt.show()
```

## Additional Matplotlib Methods & Attributes
1. `style.available` - Shows all available built-in styles for your plots

2. `style.use()` - applies a built-in style to your graph

```python
semester = [1,2,3,4,5,6]
CGPA = [7,7.5,8,8.5,9,9.8]
plt.plot(semester,CGPA,marker='.')

print(plt.style.available)
plt.style.use("dark_background")
plt.show()
```

```python
# To set back the style of graph to default
plt.style.use("default")
plt.show()
```
