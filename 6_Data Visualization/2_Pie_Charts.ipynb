### Pie Charts
- A pie chart is a circle divided into pieces to show how much each part contributes to the whole.
- Use a pie chart when:
    - You want to show how each part contributes to the whole.
    - You have categorical data
    - Categorical data has few categories (4–6); too many slices make it hard to read
